```python
files = query_data(exp_folder='/mnt/synology01/HTN0001/',
                   pattern='max_proj/<well(6)><site(3)>-<channel(3)>.<ext>',
                   plates=['401','402','403','404'],
                   plate_identifiers=['_CCU384_','_'],
                   exts=['tiff'])
files.head()
```

```python
samples = files[files['channel']=='w1'].groupby('plate').sample(n=20).reset_index(drop=True)
# samples = files[files['channel']=='ch1'].groupby('plate').sample(n=20).reset_index(drop=True)
samples
```

```python
# FFC = np.load('/mnt/synology01/HTN0001/HTN0001_ffc.npy',allow_pickle=True).item()
from skimage.morphology import remove_small_objects
VAR_THRESH = 2700
total_img = len(samples)
oof_recognized = 0
for i,sample_info in samples.iterrows():
    img = cv.imread(sample_info['file_path'],cv.IMREAD_UNCHANGED)
    # img = img/FFC['ch1']
    img[img>np.percentile(img,99.9)] = np.percentile(img,99.9)
    otsu = skimage.filters.threshold_otsu(img)
    binary = img > otsu
    binary = remove_small_objects(binary,min_size=300)
    label= skimage.measure.label(binary)
    n_cells = len(np.unique(label))-1
    img = cv.resize(img,(1080,1080),cv.INTER_AREA)
    blur_img = cv.GaussianBlur(img,(43,43),0)
    reg_var = (1-cv.Laplacian((img-img.min())/(img-img.min()+np.finfo(float).eps).max(),cv.CV_64F,ksize=9)*np.exp(-1/(n_cells+np.finfo(float).eps))).var()
    blur_var = (1-cv.Laplacian((blur_img-blur_img.min())/(blur_img-blur_img.min()).max(),cv.CV_64F,ksize=9)*np.exp(-1/(n_cells+np.finfo(float).eps))).var()
    if reg_var>VAR_THRESH and blur_var<VAR_THRESH:
        oof_recognized+=1

print('accuracy:',oof_recognized/total_img)
```
