```python
import pandas as pd
import numpy as np
import os
import query_functions_local as qfl
```

```python
data_dir = 'tests/sample_data/rxrx2/'
csv_file = os.path.join(data_dir,'test_query_data_expected_output.csv')
qfl.check_if_file_exists(csv_file=csv_file,wells=np.array(['L18']),fields=np.array(['s1','s2','s3']))
```

```python
# generate fake coords
files_df = pd.read_csv(csv_file)

for i,row in files_df.iterrows():
    img_df = pd.DataFrame(columns=list(files_df.columns)+['CoordX','CoordY'])
    img_df['cell_id'] = np.arange(0,10)
    img_df['CoordX'] = np.random.randint(540,size=10)
    img_df['CoordY'] = np.random.randint(540,size=10)
    img_df[['plate','well','site','channel','plane','filename']] =  [*row[['plate','well','site','channel','plane','filename']]]
    if i==0:
        coords_df = img_df
    else:
        print(coords_df.columns)
        coords_df = pd.concat([coords_df,img_df],ignore_index=True)

coords_df.to_csv(os.path.join(data_dir,'test_coords.csv'))
```
